```python
config_dir = '/global/homes/o/ocamacho/pydfCSR_sims/config_files/';

def modify_yaml_file(particle_group_h5, lattice_input_file, write_name, output_file):
    # Read the original YAML file
    with open(config_dir + 'W_chicane_config_base.yaml', 'r') as file:
        data = yaml.safe_load(file)
    
    # Modify the specified fields
    data['input_beam']['ParticleGroup_h5'] = particle_group_h5
    data['input_lattice']['lattice_input_file'] = lattice_input_file
    data['CSR_computation']['write_name'] = write_name
    
    # Write the modified data to a new file
    with open(output_file, 'w') as file:
        yaml.dump(data, file, default_flow_style=False)
    
    print(f"Modified file saved as: {output_file}")
```

```python
%%time
i = 0;
for file in os.listdir(init_beams_dir):
    if not file.endswith('_pg_.h5') and file.startswith('BEGBC20'):
        print(file)
        tokens = file.split('_');
        phi1 = (tokens[2])
        phi2 = (tokens[3].split('.h5')[0])
        config_name = 'W_chicane_config_base_' + phi1 + '_' + phi2 + '.yaml'
        config_path = config_dir + config_name;
        particle_group_path = init_beams_dir + file;
        output_name = 'b1le0_' + phi1 + '_' + phi2;#'w_chicane_out_' + phi1 + '_' + phi2;
        #print(particle_group_path)
        #print(config_path)
        #print(output_name)
        modify_yaml_file(particle_group_path, lattice_input_file, output_name, config_name)
        i+=1;
        ;
print(f"{i} total config files");
```
